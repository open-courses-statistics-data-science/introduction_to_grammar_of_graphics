# Using grammar of graphics approaches to visualise Gapminder data

In this problem set, we are going to visualise a dataset from Gapminder, which contains data on life expectancy, GDP per capita and population by country. For more information, see http://www.gapminder.org/data/.

Load the `gapminder.csv` data using pandas.

```python
import numpy as np
import pandas as pd
from plotnine import *
import matplotlib.pyplot as plt

df = pd.read_csv("../data/gapminder.csv")
```

Examine the first 10 rows of your dataset.

```python
df.iloc[:10, :]
```

Using Matplotlib, plot life expectancy over time for the United Kingdom.

```python
mask_uk = df["country"] == "United Kingdom"
df_uk = df[mask_uk]

plt.plot(df_uk["year"], df_uk["lifeExp"])
plt.xlabel("year")
plt.ylabel("life expectancy, years")
plt.show()
```

Again, using Matplotlib, overlay on the same plot, life expectancy plots for the UK and Burkina Faso. Add a legend which indicates which plot corresponds to which country.

```python
mask_bf = df["country"] == "Burkina Faso"
df_bf = df[mask_bf]

plt.plot(df_uk["year"], df_uk["lifeExp"], label="UK")
plt.plot(df_bf["year"], df_bf["lifeExp"], label="Burkina Faso")
plt.xlabel("year")
plt.ylabel("life expectancy, years")
plt.legend()
plt.show()
```

Now, we are going to use `plotnine`'s `ggplot` function to produce the same plot: life expectancy over time for both the UK and Burkina Faso. We first need to create a data frame that keeps only those rows for the UK and Burkina Faso.

```python
df_uk_bf = df.query('country in ["United Kingdom", "Burkina Faso"]')
```

Next run the following code to plot the life expectancies over time:

`(ggplot(df_uk_bf, aes(x='year', y='lifeExp', colour='country')) +
geom_line())`

```python
(ggplot(df_uk_bf, aes(x='year', y='lifeExp', colour='country')) +
geom_line())
```

Now change `geom_line` to `geom_point` and examine how that changes the plot.

```python
(ggplot(df_uk_bf, aes(x='year', y='lifeExp', colour='country')) +
geom_point())
```

We can also layer geoms. Now add back in `+ geom_line()` in addition to `geom_point()`. What does this plot look like?

```python
(ggplot(df_uk_bf, aes(x='year', y='lifeExp', colour='country')) +
geom_point() +
geom_line())
```

Suppose we want to add linear regression lines to data from each country. We can do this by adding `+ geom_smooth(method="lm")` to the end of our plot command.

```python
(ggplot(df_uk_bf, aes(x='year', y='lifeExp', colour='country')) +
geom_point() +
geom_line() +
geom_smooth(method="lm"))
```

We can remove the uncertainty intervals from the plots by changing `+ geom_smooth(method="lm")` to `+ geom_smooth(method="lm", se=False)`. Try this.

```python
(ggplot(df_uk_bf, aes(x='year', y='lifeExp', colour='country')) +
geom_point() +
geom_line() +
geom_smooth(method="lm", se=False))
```

Suppose instead of plotting two regression lines, we want a single regression line representing the trends across both countries. To do this, we can run the following code:

`(ggplot(df_uk_bf, aes(x='year', y='lifeExp')) +
geom_point(aes(colour='country')) +
geom_smooth(method="lm", se=False))`

```python
(ggplot(df_uk_bf, aes(x='year', y='lifeExp')) +
geom_point(aes(colour='country')) +
geom_smooth(method="lm", se=False))
```

Why has the above plotted a single regression line?

It's because whilst we use a colour aesthetic, the colour aesthetic is used only by `geom_point` and not inherited by `geom_smooth`. Whereas in the former case the colour `aesthetic` is stated in the `ggplot` section which means it is inherited by downstream geoms, including `geom_smooth` meaning it creates two regression lines  one for each colour.

Now use to `ggplot` to create a similar plot except plotting all countries in the Americas.

```python
df_americas = df.query('continent == "Americas"')

(ggplot(df_americas, aes(x='year', y='lifeExp')) +
geom_point(aes(colour='country')) +
geom_smooth(method="lm", se=False)) 
```

We can also change our straight regression line into a curvy line using a locally weighted regression (known as a `loess` line). To do this, change from:

`geom_smooth(method="lm", se=False)` to `geom_smooth()`

Try this.

```python
(ggplot(df_americas, aes(x='year', y='lifeExp')) +
geom_point(aes(colour='country')) +
geom_smooth()) 
```

We can also use `ggplot` to plot life expectancy over time for each continent in a given panel. To do so, run the following code:

`(ggplot(df, aes(x='year', y='lifeExp')) +
geom_point() +
facet_wrap('continent'))`

```python
(ggplot(df, aes(x='year', y='lifeExp')) +
geom_point() +
facet_wrap('continent')) 
```

Add a loess regression line to each of the panels above. Colour all of the regression lines orange.

```python
(ggplot(df, aes(x='year', y='lifeExp')) +
geom_point() +
geom_smooth(colour="orange") +
facet_wrap('continent')) 
```

## Open ended question

Using visualisations, assess the following statement:

"Increases in gdp per capita are associated with increases in life expectancy."

What follows is one attempt (my attempt) to examine this. Note, I am deliberately going to include all steps I take along the way to arrive at good visualisation(s) because finding these takes work! With many wrong turns along the way.

One way to assess this statement is to examine correlations between life expectancy and gdp per capita. To do this, we can first correlate these for all the data.

```python
(ggplot(df, aes(x='gdpPercap', y='lifeExp')) +
geom_point()) 
```

The above is immediately interesting: it suggests that the rate of increase of life expectancy associated with changes in income is greatest for those countries with lowest income.

Another way to plot this is using a log x scale to visualise the data.

```python
(ggplot(df, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10()) 
```

Mmm, I'm not sure I prefer the above to the non-logged graph...I may go back to the non-logged version later.

For now, I'd like to examine the relationship between income and life expectancy for a few countries because I feel like aggregating the data (as we are currently doing) may obscure important heterogeneities. To start, I am going to look at the UK, Burkina Faso, Brazil and Norway: all countries at various points along the spectrum.

```python
df_selection = df.query('country in ["United Kingdom", "Burkina Faso", "Brazil", "Norway"]')

(ggplot(df_selection, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10()) 
```

Ok, the above isn't too helpful, as I've not added in country identifiers.

```python
(ggplot(df_selection, aes(x='gdpPercap', y='lifeExp', colour='country')) +
geom_point() +
scale_x_log10()) 
```

Ok, the above is interesting but it's a bit difficult to see the relationships when they're all plotted on the same graph. So, I'm going to create panels for each graph.

```python
(ggplot(df_selection, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10() +
facet_wrap('country')) 
```

Now, I am going to add linear regression lines to try to gauge the differences in association among this selection.

```python
(ggplot(df_selection, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10() +
geom_smooth(method="lm", se=False, colour="orange") +
facet_wrap('country')) 
```

The above suggests a few things: a) that the association between income and life expectancy varies a great deal between countries and b) it hints that, as income increases the association decreases.

Let's now pick a larger sample: all the countries in Europe. What does the relationship look like?

```python
df_europe = df.query('continent == "Europe"')

(ggplot(df_europe, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10() +
geom_smooth(method="lm", se=False, colour="orange") +
facet_wrap('country')) 
```

Hmm, this is ok and would probably work if I exported the figure as a big graph.

An alternative way to look at this is to visualise the regression lines only.

```python
(ggplot(df_europe, aes(x='gdpPercap', y='lifeExp', colour='country')) +
scale_x_log10() +
geom_smooth(method="lm", se=False)) 
```

Kind of interesting...but still not sure it's be best plot.

Ok, let's go back to our previous plot. Instead of focussing on Europe, let's plot a selection of countries from each range of income quantiles.

```python
# determine mean incomes of each country and sort data frame according to these
df_sorted = df.groupby('country').mean().sort_values(by='gdpPercap').reset_index()
```

First let's look at the bottom 20 countries by mean income.

```python
bottom_20 = list(df_sorted["country"][:20])
df_few = df.query('country in @bottom_20')

(ggplot(df_few, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10() +
geom_smooth(method="lm", se=False, colour="orange") +
facet_wrap('country')) 
```

Interesting: lots of heterogeneity. Let's now reorder according to income and see if any patterns emerge.

```python
df_few["country"] = df_few["country"].astype('category')
df_few["country"] = df_few["country"].cat.reorder_categories(bottom_20)

(ggplot(df_few, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10() +
geom_smooth(method="lm", se=False, colour="orange") +
facet_wrap('country')) 
```

Still there's a load of heterogeneity: it seems that increases in income at this end of the scale seem to potentially have a variety of associations with life expectancy. Could this be because there are many different ways for income to increase? For example, finding natural resources versus building a booming export economy? These different causes of changes in income likely mean that the relationship with life expectancy is quite various.

How about the top 50 instead?

```python
upper_index = 20
top = list(df_sorted["country"][-upper_index:])
df_few = df.query('country in @top')

(ggplot(df_few, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10() +
geom_smooth(method="lm", se=False, colour="orange") +
facet_wrap('country')) 

df_few["country"] = df_few["country"].astype('category')
df_few["country"] = df_few["country"].cat.reorder_categories(top)

(ggplot(df_few, aes(x='gdpPercap', y='lifeExp')) +
geom_point() +
scale_x_log10() +
geom_smooth(method="lm", se=False, colour="orange") +
facet_wrap('country')) 
```

Interesting. These relationships (with the exception of the OPEC countries) are fairly stable and suggest that, as income increases, there is a decreased association between income and life expectancy.
